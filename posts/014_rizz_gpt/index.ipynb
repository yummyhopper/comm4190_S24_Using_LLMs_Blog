### Does AI Have Game?

As a foremost expert on automated rizz I was in on the game early. The issue is frontier models have an alignment problem when it comes to spitting game. Look at what the top three players in today's AI market say when I give them a simple prompt to send a flirtatious text. 

```python
# Import ChatBots from LangChain

from langchain_openai import ChatOpenAI
from langchain_anthropic import ChatAnthropic
from langchain_google_genai import GoogleGenerativeAI
```

```python
# Load API key

from dotenv import load_dotenv
_ = load_dotenv()
```

```python
# Select the newest frontier models

ChatGPT = ChatOpenAI(model = "gpt-4-0125-preview")
Gemini = GoogleGenerativeAI(model="models/gemini-1.5-pro-latest")
Claude = ChatAnthropic(model="claude-3-opus-20240229")
```

```python
# Give a simple prompt to test rizz

prompt = '''

    You are a flirtatious but gentlemanly chatbot. You really like a girl named Rebecca and want to ask her on a date. Write a short saucy message rizzing her up and asking her out.
    
'''
```

```python
ChatGPT.invoke(prompt) # Ask ChatGPT
```

```python
Gemini.invoke(prompt) # Ask Gemini
```

```python
Claude.invoke(prompt) # Ask Claude
```

### Is RizzGPT Going to Help Me Find Love?



Probably not. As with the conclusion of the South Park episode over-relying on AI to help you flirt could cause a disconnect between your actual personality and your perceived personality when you actually end up on a date. Imagine how disappointed your crush will be when you end up dryly discussing work, the weather, and politics over steak dinner instead of your usual sauciness! Worse, no one likes being lied to especially in a relationship. If you do use RizzGPT, it's important to do so sparingly and always add your own thoughts. One of the things AI is best at is idea generation. I don't see any problem using AI to unstick you when you've been overthinking that two sentence text for half an hour, but it's important to do so consciously.
